```python
class GradeManager:
    def __init__(self):
        self.records = {}

    def add_student(self, name):
        if name not in self.records:
            self.records[name] = {}
        else:
            print(f"{name} already exists.")

    def add_subject(self, name, subject, grade):
        if name in self.records:
            self.records[name][subject] = grade
        else:
            print(f"Student {name} not found.")

    def remove_subject(self, name, subject):
        if name in self.records and subject in self.records[name]:
            del self.records[name][subject]
        else:
            print(f"Subject {subject} not found for {name}.")

    def get_average(self, name):
        if name in self.records and self.records[name]:
            grades = list(self.records[name].values())
            return sum(grades) / len(grades)
        else:
            return None

    def display_all(self):
        for name, subjects in self.records.items():
            avg = self.get_average(name)
            print(f"\nStudent: {name}")
            for subject, grade in subjects.items():
                print(f"  {subject}: {grade}")
            if avg is not None:
                print(f"  Average: {avg:.2f}")


# Interactive menu
manager = GradeManager()

while True:
    print("\n--- Grade Manager ---")
    print("1. Add Student")
    print("2. Add Subject & Grade")
    print("3. Remove Subject")
    print("4. Show All Records")
    print("5. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        name = input("Enter student name: ")
        manager.add_student(name)

    elif choice == "2":
        name = input("Enter student name: ")
        subject = input("Enter subject: ")
        grade = float(input("Enter grade: "))
        manager.add_subject(name, subject, grade)

    elif choice == "3":
        name = input("Enter student name: ")
        subject = input("Enter subject to remove: ")
        manager.remove_subject(name, subject)

    elif choice == "4":
        manager.display_all()

    elif choice == "5":
        print("Exiting program...")
        break

    else:
        print("Invalid choice, try again.")
```
